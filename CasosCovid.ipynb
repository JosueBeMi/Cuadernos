```
import pandas as pd
df=pd.read_csv('23totalcases.csv')
df['Mexico'].plot(kind='line')
```

```
pd.set_option('max_rows',99999)
df=df.fillna(method='ffill')
df['Mexico']
```

```
df2=df.T
df2=df2.drop(['date'])
df2=df2.reset_index()
df2=df2.rename({'index':'Paises',83:'Total'},axis=1)
print(df2.head(5))
```

```
import plotly.express as px
fig = px.line(df2,x='Paises', y='Total')
fig.show()
```

```
import plotly.graph_objects as go
from plotly.subplots import make_subplots
cols_plot = ['China', 'Italy', 'Iran','Mexico']
fig = make_subplots(rows=len(cols_plot), cols=1, start_cell="bottom-left")
i=1
j=0
for col in cols_plot:
 trace=go.Scatter(x=df[col], y=df['date'],mode='lines',name=cols_plot[j])
 fig.append_trace(trace,i,1)
 i=i+1
 j=j+1
fig.show()
```

```
df3=df2[(df2['Total']>200)&(df2['Total']<=320)]
df3=df3.T
df3=df3.reset_index()
nombres=df3.iloc[0] 
df3=df3[1:]
df3.columns=nombres
print(list(df3))
df3=df3.drop(['Paises'],axis=1)
print(list(df3))
df3=df3.fillna(method='ffill')
df3
```

```
import cufflinks as cf
import plotly.offline as py
py.iplot([{
   'x': df3.index,
   'y': df3[col],
   'name': col
}  for col in df3.columns], filename='cufflinks/multiple-lines-on-same-chart')
```
